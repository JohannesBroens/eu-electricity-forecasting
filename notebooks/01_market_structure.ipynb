# Market Structure Reference

## Bidding Zones

DK1 (West Denmark) and DK2 (East Denmark) sit at the intersection of the Nordic and Continental synchronous areas. DK1 connects to NO2, SE3, and DE_LU; DK2 connects to SE4 and DE_LU. The Great Belt interconnector (DK1-DK2, 600 MW) is the only AC link between the two synchronous systems in Denmark.

## Day-Ahead Auction

Nord Pool operates the day-ahead auction (SDAC). Gate closure at 12:00 CET, results published ~12:42 CET. The auction clears all 24 hours of the next delivery day simultaneously via EUPHEMIA, the pan-European coupling algorithm. Prices are set by marginal pricing within each bidding zone; cross-border flows are determined by available transfer capacity and price differentials.

## Merit Order and Marginal Pricing

Generators bid at marginal cost. The clearing price equals the bid of the most expensive unit needed to meet demand (the marginal unit). Near-zero marginal cost renewables (wind, solar) push the supply curve right, displacing higher-cost thermal generation and suppressing prices -- the merit order effect.

## Danish Wind Penetration

Denmark's wind penetration reached ~57% of total electricity consumption in 2024 (Danish Energy Agency). This creates price dynamics not seen in thermal-dominated markets: frequent negative prices during high-wind/low-demand periods, large intraday price swings driven by wind forecast errors, and strong coupling to German prices via the 2500 MW DK1-DE_LU interconnector.
